```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
att_ratio = 100*np.random.random(100)
succes = 20+att_ratio**2/130 + np.random.normal(0,10,100)

N=5
mean_years_in_phd = (6.4,5.7,5.2,4.5,3.8)
ind = np.arange(N)
yerr = np.random.random(5)/10*5
width = 0.7
```

```python
fig=plt.figure()
ax1 = plt.subplot2grid((2,2),(0,0))
ax2 = plt.subplot2grid((2,2),(0,1))
ax3 = plt.subplot2grid((2,2),(1,0),colspan=2)

ax1.plot(att_ratio,succes,'o')
ax1.set_xlabel(u'Présence au club de lecture (%)')
ax1.set_ylabel(u'Succès académique(%)')
ax1.set_ylim(0,100)

ax2.bar(ind,mean_years_in_phd,width,color='seagreen',yerr=yerr)
ax2.set_xticks(ind + width/2)
ax2.set_xticklabels(('<20', '20-40', '40-60', '60-80', '>80'))
ax2.set_xlabel(u'Présence au club de lecture (%)')
ax2.set_ylabel(u'Années \n avant graduation')

labels = ':\'(', ':o', ':(', ':S'
sizes = [15, 30, 45, 20]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0.1, 0.1, 0.2, 0.1)

ax3.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.0f%%', shadow=False, startangle=90)
ax3.set_xlabel(u'Réaction de ton directeur \n lors de ton absence \n')    
ax3.axis('equal')

fig.tight_layout();
```
